````python
!pip install pytube
from pytube import YouTube
import os

def downloadYouTube(videourl, path):

    yt = YouTube(videourl)
    yt = yt.streams.filter(progressive=True, file_extension='mp4').order_by('resolution').desc().first()
    if not os.path.exists(path):
        os.makedirs(path)
    yt.download(path)

downloadYouTube("https://www.youtube.com/watch?v=mslTnVx6VPQ", './')
````

## Now this is the time for implementing segmentation

````python
from skimage.io import *
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure, imshow, axis


image = imread("https://raw.githubusercontent.com/ayoolaolafenwa/PixelLib/master/Images/github.jpg")
imsave('image.jpg',image)
figure(figsize = (20,20))
imshow(image, interpolation='nearest')
axis('off')
````

````python
  %%capture
  import pixellib
  from pixellib.torchbackend.instance import instanceSegmentation

  ins = instanceSegmentation()
  ins.load_model("pointrend_resnet50.pkl")
  ins.segmentImage("image.jpg", show_bboxes=True, output_image_name="output_image.jpg")
````

## Visualisation of the segmented image result.

````python
img=imread("output_image.jpg")
figure(figsize = (20,20))
imshow(img, interpolation='nearest')
axis('off')
````
