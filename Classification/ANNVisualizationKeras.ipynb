```python
Hidden_Layers = 4 #@param {type:"slider", min:1, max:4, step:1}
Optimizer = "rmsprop" #@param ["adam", "rmsprop"]
Loss = 'binary_crossentropy' #@param ["categorical_crossentropy", "binary_crossentropy"]
DropOut = 0.3 #@param {type: "slider", min:0.1, max :0.5, step: 0.1}
Activation = "softmax" #@param ["sigmoid", "softmax"]
File_Name = 'Cancer Classification' #@param {type:"string"}
Batch_Size = 23 #@param {type: "slider", min:10, max :40, step: 1}
Epochs = 500 #@param {type: "slider", min:100, max:500, step: 100}
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

model = Sequential()
#First layer
model.add(Dense(activation = Activation, input_dim = 29, units = 16, kernel_initializer = "uniform")) # input
model.add(Dropout(0.5)) # Hidden unit
model.add(Dense(activation = Activation, units = 1, kernel_initializer = "uniform")) # OutPut

model.summary()
```

## **`Visualkeras & visualizer ` ** Visualizing Keras model

```python
import visualkeras
from PIL import ImageFont

#visualkeras.layered_view(model, legend=True) # without custom font
font = ImageFont.truetype("Humor-Sans.ttf", 12)
visualkeras.layered_view(model, legend=True, font=font) # selected font
```
