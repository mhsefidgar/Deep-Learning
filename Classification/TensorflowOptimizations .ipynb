#Deep learning optimizations 
##Importing necessary libraries

```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from tensorflow.keras.datasets import *
from tensorflow.keras.optimizers import *
```

##Create a sample CNN and loading mnist data

```python
model= Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128,activation='relu'),
    Dropout(0.2),
    Dense(10,activation='softmax')
])
(XTrain, YTrain), (XTest, YTest) = mnist.load_data()
```

##Create optimizers


*   RMSprop
*   Adam
*  Adagrad
* Adadelta




```python
sgd = SGD(learning_rate=0.01, decay=1e-6, momentum=0.0, nesterov=True)
RMS = RMSprop(learning_rate=0.001,rho=0.9,momentum=0.0,epsilon=1e-07,centered=False,name='RMSprop')
ADM = Adam(learning_rate=0.001,beta_1=0.9,beta_2=0.999,epsilon=1e-07,amsgrad=False,name='Adam')
ADG = Adagrad(learning_rate=0.001,initial_accumulator_value=0.1,epsilon=1e-07, name='Adagrad')
ADADELL=Adadelta(learning_rate=0.001,rho=0.95,epsilon=1e-07,name='Adadelta')
```

#Define the models for training and optimization 

```python
model.compile(optimizer=sgd,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
modelsgd = model;
modelRMS = model;
modelADM = model;
modelADG = model;
modelADADELL = model;
modelsgd.compile(optimizer=sgd, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
modelRMS.compile(optimizer=RMS, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
modelADM.compile(optimizer=ADM, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
modelADG.compile(optimizer=ADG, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
modelADADELL.compile(optimizer=ADADELL, loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

##Train and evaluate the developed models

```python
%%capture
historysgd = modelsgd.fit(XTrain, YTrain, epochs=5)
modelsgd.evaluate(XTest, YTest, verbose=2)
historyRMS = modelRMS.fit(XTrain, YTrain, epochs=5)
modelRMS.evaluate(XTest, YTest, verbose=2)
historyADM = modelADM.fit(XTrain, YTrain, epochs=5)
modelADM.evaluate(XTest, YTest, verbose=2)
historyADG = modelADG.fit(XTrain, YTrain, epochs=5)
modelADG.evaluate(XTest, YTest, verbose=2)
historyADADELL = modelADADELL.fit(XTrain, YTrain, epochs=5)
modelADADELL.evaluate(XTest, YTest, verbose=2)
```

```python
#fig, ax = subplots(1, 2)
#fig.suptitle('Model Accuracy and loss function diagram')
'''
historysgd
historyRMS
historyADM
historyADG
'''

figure(figsize=(15,20))
suptitle('Model Accuracy and loss function diagram')
subplot(5,2,1)
title('Model loss - SGD');
plot(historysgd.history['loss'], 'r', linewidth=4, markersize=1,label='Training Loss')
#plot(history.history['val_loss'], label='Validation Loss')
legend()

subplot(5,2,2)
title('Model accuracy - SGD');
plot(historysgd.history['accuracy'], 'b', linewidth=4, markersize=1, label='Training Accuracy')
#plot(history.history['val_accuracy'], label='Validation Accuracy')
legend()

subplot(5,2,3)
title('Model loss - RMSProp');
plot(historyRMS.history['loss'], 'r', linewidth=4, markersize=1,label='Training Loss')
#plot(history.history['val_loss'], label='Validation Loss')
legend()

subplot(5,2,4)
title('Model accuracy - RMSProp');
plot(historyRMS.history['accuracy'], 'b', linewidth=4, markersize=1, label='Training Accuracy')
#plot(history.history['val_accuracy'], label='Validation Accuracy')
legend()

subplot(5,2,5)
title('Model loss - Adam');
plot(historyADM.history['loss'], 'r', linewidth=4, markersize=1,label='Training Loss')
#plot(history.history['val_loss'], label='Validation Loss')
legend()

subplot(5,2,6)
title('Model accuracy - Adam');
plot(historyADM.history['accuracy'], 'b', linewidth=4, markersize=1, label='Training Accuracy')
#plot(history.history['val_accuracy'], label='Validation Accuracy')
legend()

subplot(5,2,7)
title('Model loss-Adagrad');
plot(historyADG.history['loss'], 'r', linewidth=4, markersize=1,label='Training Loss')
#plot(history.history['val_loss'], label='Validation Loss')
legend()

subplot(5,2,8)
title('Model accuracy-Adagrad');
plot(historyADG.history['accuracy'], 'b', linewidth=4, markersize=1, label='Training Accuracy')
#plot(history.history['val_accuracy'], label='Validation Accuracy')
legend()

subplot(5,2,9)
title('Model loss-Adadelta');
plot(historyADADELL.history['loss'], 'r', linewidth=4, markersize=1,label='Training Loss')
#plot(history.history['val_loss'], label='Validation Loss')
legend()

subplot(5,2,10)
title('Model accuracy-Adadelta');
plot(historyADADELL.history['accuracy'], 'b', linewidth=4, markersize=1, label='Training Accuracy')
#plot(history.history['val_accuracy'], label='Validation Accuracy')
legend()

show()
```
